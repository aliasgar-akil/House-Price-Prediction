# NA Filling

```python
na_list = list(df_train.isna().sum()[df_train.isna().sum() > 0].index)
na_list
```

```python
for col in na_list:
    df_train[col].fillna(value = df_train[col].median(), inplace=True)
    df_test[col].fillna(value = df_train[col].median(), inplace=True)
```

# Feature Encoding

```python
df = pd.concat([df_train,df_test],axis=0)
df.shape
```

```python
columns_numeric = list(df.dtypes[(df.dtypes == 'int64') | (df.dtypes == 'float64')].index)
columns_object = list(df.dtypes[df.dtypes == 'object'].index)
columns_object
```

```python
df = pd.get_dummies(data = df, columns = columns_object, drop_first = True)
df
```

```python
df = df[list(df.corr()['SalePrice'][(df.corr()['SalePrice'] > 0.3) | (df.corr()['SalePrice'] < -0.3)].index)]
```

# Model

```python
X_train = df.iloc[:1352].drop('SalePrice',axis=1)
y_train = df.iloc[:1352]['SalePrice']
X_test = df.iloc[1352:].drop('SalePrice',axis=1)
X_train.shape,X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model_l = LinearRegression()
model_l.fit(X_train,y_train)
model_l.score(X_train,y_train)
```

```python
df_submit = pd.DataFrame({'Id':Id, 'SalePrice':model_l.predict(X_test)})
df_submit.head()
```

```python
df_submit.to_csv('submit_01',index=False)
```

```python
from sklearn.ensemble import RandomForestRegressor
model_R = RandomForestRegressor(n_estimators=800,max_depth=11)
model_R.fit(X_train,y_train)
model_R.score(X_train,y_train)
```

```python
df_submit = pd.DataFrame({'Id':Id, 'SalePrice':model_R.predict(X_test)})
df_submit.to_csv('submit_03',index=False)
```
